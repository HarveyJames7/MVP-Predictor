```python
# This is where I will be cleaning the data 
import pandas as pd
```

```python
# Load in mvps csv file
mvps = pd.read_csv("nbaMvps.csv")
```

```python
mvps
```

```python
# Only selecting the columns that we need from the dataset to avoid duplicates
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.head()
```

```python
# Load in the players csv file
players = pd.read_csv("nbaPlayers.csv")
```

```python
players.head()
```

```python
# Delete the unneeded columns
del players["Unnamed: 0"]
del players["Rk"]
```

```python
# Fix the player's names. We need to remove the * when they appear
players["Player"] = players["Player"].str.replace("*", "", regex = False)
```

```python
players.head(20)
```

```python
# Here you can see that players who played for several teams will appear multiple times
players.groupby(["Player", "Year"])
```

```python
# Here I designed a function to only keep the last team a player played for if he was on several teams
# This way the total stats for each player are kept and there are no duplicates
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
# The previous function added 2 extra index columns (Player and Year) which we do not want
players.head(20)
```

```python
# Drop index level
players.index = players.index.droplevel()
```

```python
# Drop index level
players.index = players.index.droplevel()
```

```python
# Here we can see that the index levels were dropped along with the duplicates of players in the same year
players[players["Player"] == "Greg Anderson"]
```

```python
# Here we combine the player and mvp data
combined = players.merge(mvps, how = "outer", on = ["Player", "Year"])
```

```python
combined.head(5)
```

```python
# This is to view only players who won MVP pts
combined[combined["Pts Won"] > 0]
```

```python
# Here we replace 'NaN' with 0 for pts won, pts max, and share
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined
```

```python
# Load in the teams csv file
teams = pd.read_csv("nbaTeams.csv")
```

```python
teams.head(5)
```

```python
# Remove the division rows
teams = teams [~teams["W"].str.contains("Division")]
```

```python
# Remove the * from teams
teams["Team"] = teams ["Team"].str.replace("*", "", regex = False)
```

```python
# There are full team names and abbreviated team names which is a problem
# Here is the full team names
teams["Team"].unique()
```

```python
# Here is the abbreviated team names
combined["Tm"].unique()
```

```python
# Here is where we will create a dictionary to map between the full names and abbreviations 
nicknames = {}

with open ("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        na, abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
# Here we combined Tm and Team
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
# Here we combine the combined dataframe with the teams dataframe
stats = combined.merge(teams, how = "outer", on = ["Team", "Year"])
```

```python
stats
```

```python
# The number of rows is the same as stats
combined
```

```python
# We do not want the Unnamed column
del stats["Unnamed: 0"]
```

```python
# Here we try to convert all of the columns in stats to numeric if possible
stats = stats.apply(pd.to_numeric, errors = "ignore")
```

```python
stats.dtypes
```

```python
# The GB section has a '-' in it which means that a team is 0 games behind
stats["GB"].unique()
```

```python
# Replace '-' with 0
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
# Convert GB to numeric
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats = stats.drop(18285)
```

```python
# Write stats to csv
stats.to_csv("nbaPlayerMvpStats.csv")
```

```python
# Here we want to see who scored the most points in a season
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending = False).head(10)
```
